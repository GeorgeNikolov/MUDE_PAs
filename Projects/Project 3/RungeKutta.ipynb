## Overview

The Runge-Kutta methods were developed around 1900 by Carl Runge and Wilhelm Kutta as a family of implicit and explicit methods for the approximate solutions of time-dependent simultaneous non-linear differential equations.

They are a 'family' of methods because they have a common underlying process using which, one can derive a Runge-Kutta method of any order. All one needs, is an initial value problem i.e. the time-derivative of a continuous function and the value of that function at the first time step.

## Part 1

Consider the initial value problem

$$\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0$$

The time step is $\Delta t$. A $p$-th order Runge-Kutta method requires $p$ stages:

$$ k_1 = f(t_n, y_n) $$ 
$$ k_2 = f(t_n + c_2 \Delta t, y_n + (a_{21}k_1)\Delta t) $$
$$ k_3 = f(t_n + c_3 \Delta t, y_n + (a_{31}k_1 + a_{32}k_2)\Delta t) $$
$$ k_4 = f(t_n + c_4 \Delta t, y_n + (a_{41}k_1 + a_{42}k_2 + a_{43}k_3)\Delta t) $$
.... 
$$k_p = f(t_n + c_p \Delta t, y_n + (a_{p1}k_1 + a_{p2}k_2 + a_{p3}k_3 + ....+ a_{p,p-1}k_{p-1})\Delta t)$$

A method of the Runge-Kutta family can then be expressed in parametric form as:

$$y_{n+1} = y_n + \Delta t \sum_{i=1}^{p} b_i k_i + O(h^{p+1})$$

Use the series above to write down the <b>slopes $(k)$ </b> and the parametric form of the 2nd order explicit Runge-Kutta method $(p = 2)$. You can leave the parameters $a_{ij}$, $b_i$, and $c_i$ as they are.

Logically, there are infinite combinations of these parameters, i.e., $a_{ij}$, $b_i$, and $c_i$. But for a set of parameters that delivers the desired order of accuracy consistently, the Runge-Kutta system requires these parameters to be consistent with the Taylor series expansion of $y_{t+\Delta t}$.

## Error Analysis

The code provides you the arrays of $t$ and $y$ for the various methods and the exact solution at the values specified in $t$. The following formulas describe how one can calculate the overall error.

$$e_n = \left(\frac{1}{N} \sum_{i=1}^{N} (\epsilon_i)^n \right)^{\frac{1}{n}}$$


$$e_\infty = |\epsilon_i|_{\text{max}}$$

Determine $e_2$ and $e_\infty$ at each time-step for each method from each of the $dt$ values. 

$\epsilon_i$ is the difference between the values of $y$ numerically at the $i$-th time-step and the value as per the exact analytical solution at that time-step. Whereas, $e_\infty$ is the maximum value of $\epsilon_i$ within the time domain. $e_n$ is the $n$-th norm given by the upper equation. If $n=1$ then $e_n$ is just the artimetic mean of all $\epsilon$. $N$ is the total number of time steps, which is the same as the number of $\epsilon$ values.

```python
import pandas as pd
```

```python
dt = [0.00125, 0.0025, 0.005, 0.01, 2/100]

def error_2(dt):
    errorlijst = []
    for i in range(len(dt)):
        T = 10
        N = round((T/dt[i])/10) * 10    #Rounds to the nearest 10
        # Time array
        t = np.linspace(0, T, N + 1)
        solFE = forward_euler(f, t, y0 = 0)
        solBE = backward_euler(f, dfdx, t, y0 = 0, max_iter=100, tol=1e-6)
        solRK4 = runge_kutta_4(f, t, y0 = 0)
        solRK2 = runge_kutta_2(f, t, y0 = 0)
        sol_exact = exact(t)
        methods = [solFE, solRK2, solRK4, solBE]
        error2 = []
        errorlijst.append(error2)
        for i in range(len(methods)):
            error_i = sol_exact - methods[i]
            err2 = (1/N * np.sum((error_i)**2))**(1/2)
            error2.append(err2)
        
    return errorlijst

def error_inf(dt):
    errorlijst = []
    for i in range(len(dt)):
        T = 10
        N = round((T/dt[i])/10) * 10    #Rounds to the nearest 10
        # Time array
        t = np.linspace(0, T, N + 1)
        solFE = forward_euler(f, t, y0 = 0)
        solBE = backward_euler(f, dfdx, t, y0 = 0, max_iter=100, tol=1e-6)
        solRK4 = runge_kutta_4(f, t, y0 = 0)
        solRK2 = runge_kutta_2(f, t, y0 = 0)
        sol_exact = exact(t)
        methods = [solFE, solRK2, solRK4, solBE]
        errorinf = []
        errorlijst.append(errorinf)
        for i in range(len(methods)):
            error_i = sol_exact - methods[i]
            errinf = np.max(np.abs(error_i))
            errorinf.append(errinf)
        
    return errorlijst
    
    
```

True or False: The backward Euler (`BE`) is inherently stable and hence, the maximum error is bounded. Hint: Try to increase the time step dt to a larger value.

</p>
</div>

It is true. Because the maximum timestep for BE is limited by: 

$$|\frac{1}{1+\alpha h}| < 1$$ 

where in our case 

$$\alpha = 100$$

, so with an increasing time step the denominator will only grow bigger, resulting in a smaller result and thus always being smaller than 1.
